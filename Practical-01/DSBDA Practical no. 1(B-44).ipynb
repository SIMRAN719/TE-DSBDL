```python
#calculate the mean of temperature column
mean_value=df['temperature'].mean() 
print(mean_value)
```

```python
df['temperature']=df['temperature'].fillna(mean_value) #replace NaN value with mean value
```

```python
print(df)
```

```python
df.dropna(subset=['windspeed']) #delete NaN value which is contained by windspeed
```

```python
mean_value=df['windspeed'].mean() #calculate the mean of windspeed column
print(mean_value)
```

```python
df['windspeed']=df['windspeed'].fillna('no_windspeed') #replace NaN value with no_windspeed
print(df)
```

```python
df.dropna(subset=['event']) #delete NaN value which is contained by event
```

```python
df['event']=df['event'].fillna('no_event') #replace NaN value with no_event
print(df)
```
