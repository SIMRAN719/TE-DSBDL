Distributional plots, as the name suggests are type of plots that show the statistical distribution of data. The distplot() shows the histogram distribution of data for a single
column. The column name is passed as a parameter to the distplot() function. Let's see how the price of the ticket for each passenger is distributed

__Distribution Plots__

a. Distplot

b. jointplot

c. Pairplot

d. Rugplot

* These plots help us to visualize the distribution of data. We can use these plots to understand the mean, median, range, variance, deviation, etc of the data.

* Distplot: Distplot gives us the histogram of the selected continuous variable.
It is an example of a univariate analysis.

(Univariate analysis is a statistical method used to analyze and describe the distribution, central tendency, and dispersion of one variable at a time. It focuses on understanding the characteristics and properties of individual variables independently of other variables. )

* We can change the number of bins i.e. number of vertical bars in a histogram

```python
#Distplot
sns.distplot(dataset['Fare'])
```

```python
sns.displot(dataset['Fare'])
```

```python
#Kernal density estimation
#You can also pass the value for the bins parameter in order to see more or less details in the graph.
sns.distplot(dataset['Fare'], kde=False, bins=10)
```

```python
#Jointplot
sns.jointplot(x='Age', y='Fare', data=dataset)
```

```python
#Pairplot
sns.pairplot(data=dataset)
```

```python
sns.pairplot(data=dataset, hue='Sex')
```

```python
#RugPlot
sns.rugplot(dataset['Fare'])
```

```python
sns.barplot(x='Age', y='Fare', data=dataset)
```

```python
sns.barplot(x='Sex', y='Age', data=dataset)
```

```python
sns.countplot(x='Sex', data=dataset)
```

```python
sns.boxplot(x='Sex', y='Age', data=dataset)
```

```python
sns.boxplot(x='Sex', y='Age', data=dataset, hue="Survived")
```

```python
sns.violinplot(x='Sex', y='Age', data=dataset)
```

```python
sns.stripplot(y = dataset['Age'], x = dataset['Pclass'])
```

```python
sns.swarmplot(y = dataset['Age'], x = dataset['Pclass'])
```
